```python
### importing numpy pandas and seaborn for analysis and visualisation

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
###uploading dataset using read_csv function in pandas and then printing first  5 row using pandas 's head function

df_tracks = pd.read_csv('C:/Users/ranji/OneDrive/Documents/Desktop/dataAnalytic/tracks.csv')
df_tracks.head()
```

```python
###finding number of null values in each column using 
###isnull function and sum function

pd.isnull(df_tracks).sum()
```

```python
###from above output its clear that 71 row of column name is null
##now we will be using info function of pandas to get an idea about how much row ,column , datatype ,memory usases are there
df_tracks.info()
```

```python
#here we can see there are total 586672 row for each column except name,obviously 71 less
###finding top 10 least poppular song using sort_values function
sorted_df = df_tracks.sort_values('popularity' , ascending = True).head(10)
sorted_df
```

```python
### to get descroptive statistics about the data using describe function like mean mode  std max etc
df_tracks.describe().transpose()
```

```python
###finding top 10 most popular song using query function
most_popular = df_tracks.query('popularity> 90',inplace= False).sort_values('popularity' , ascending=False)
most_popular[:10]
```

```python
###setting index to 'release-date' using set_index function then convert to datetime 
df_tracks.set_index("release_date" , inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks[:10]
```

```python
###query for a particular element using index and name lets say artist of column number 18
df_tracks[['artists']].iloc[18]
```

```python
###changing duration from millisecond to second as nearest integer using lambda function and make these changes to original data as well 
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x : round(x/1000))
df_tracks.drop("duration_ms" , inplace = True, axis = 1)
df_tracks.duration.head()
```

```python
###first visualisation heat map using sns without unwanted columns 'key','mode' , 'explict'
###storing data into a new variable, not updating changes into my original dataset
corr_df = df_tracks.drop(['key','mode','explicit'], axis = 1).corr(method = "pearson")
###setting figsize
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(corr_df,annot = True,fmt = ".1g" , vmin=-1,center=0,vmax=1,cmap="inferno",linewidth=1,linecolor="Black")
###-1 is negative correlation while 1 is possitive relation
heatmap.set_title("Correlation HeatMap between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
###creating some small sample data from original data 0.4% of original data
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
###checking if sample data created is right in temp of length of row
print(len(sample_df))
```

```python
###ploting regression for loudness and energy
plt.figure(figsize= (10,6))
sns.regplot(data=sample_df , y = "loudness" , x = "energy" ,color = "c").set(title="Correlation between Loudness and Energy")
```

```python
###checking correlation between  popularity and acousticness
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y="popularity" , x = "acousticness" , color ="b").set(title = "Polularity vs Acousticnes")
```

```python
###from above two correlation plot its clear that in first plt loudness and energy have possitive correlation while in second regpression plt
###the correltion between popularity and acousticness
```

```python
#creating a new column 'year' from 'released_date'
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
###checking version of seaborn
print(sns.__version__)
```

```python
###ploting distribution plot to know number of song released each year since 1922 to 2021
sns.displot(years , discrete=True  ,aspect =2 , height=5, kind="hist").set(title="Number of song Released per year")
```

```python
###bar graph to see how duration changes each year
total_dr=df_tracks.duration
fig_dim = (18,7)
fig,ax = plt.subplots(figsize = fig_dim)
fig = sns.barplot(x=years , y=total_dr , ax=ax, errwidth =False).set(title="Years vs durations")
plt.xticks(rotation =90)
```

```python
###analysing average duration of song each year
###bar graph to see how duration changes each year
total_dr=df_tracks.duration
sns.set_style(style="whitegrid")
fig_dim = (10,5)
fig,ax = plt.subplots(figsize = fig_dim)
fig = sns.lineplot(x=years , y=total_dr , ax=ax, ).set(title="Years vs durations")
plt.xticks(rotation =60)
```
